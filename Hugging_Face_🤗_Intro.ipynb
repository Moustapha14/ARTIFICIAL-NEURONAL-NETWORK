En intelligence artificielle (IA), un modèle est une représentation mathématique d'un système ou d'un processus. Plus précisément, un modèle en IA est une fonction mathématique ou statistique qui est entraînée sur des données pour effectuer une tâche spécifique.

Un modèle en IA peut être de différents types, selon la tâche à accomplir. Voici quelques exemples de types de modèles en IA :

1. **Réseaux de neurones artificiels (ANN) :** Ce sont des modèles inspirés par le fonctionnement du cerveau humain. Ils sont constitués de couches de neurones interconnectés et sont utilisés pour des tâches telles que la classification d'images, la traduction automatique, etc.

2. **Réseaux de neurones convolutifs (CNN) :** Ce sont des types spécifiques de réseaux de neurones adaptés au traitement d'images. Ils sont très efficaces pour extraire des caractéristiques à partir d'images.

3. **Réseaux de neurones récurrents (RNN) :** Ils sont conçus pour traiter des données séquentielles, comme du texte ou de l'audio. Les RNN sont souvent utilisés pour la génération de texte, la traduction automatique, etc.

4. **Transformers :** Les transformers sont des architectures de réseau neuronal qui se sont révélées très efficaces pour de nombreuses tâches en traitement du langage naturel (NLP), telles que la classification de texte, la génération de texte, etc.

5. **Modèles probabilistes :** Ces modèles utilisent des concepts de probabilité pour prendre des décisions. Par exemple, les modèles de Markov cachés (HMM) sont souvent utilisés pour la reconnaissance de la parole et la reconnaissance de l'écriture manuscrite.

En résumé, un modèle en IA est un outil mathématique qui apprend à partir de données et qui est capable d'effectuer des tâches spécifiques, telles que la reconnaissance d'images, la traduction de texte, la prise de décision, etc.



Dans le domaine du traitement du langage naturel (NLP), un modèle est une représentation mathématique d'un système capable de comprendre et de générer du langage humain. Ces modèles sont entraînés sur de grandes quantités de données textuelles pour effectuer des tâches telles que la classification de texte, la génération de texte, la traduction automatique, etc.

Voici quelques types de modèles couramment utilisés en NLP :

1. **Modèles de sacs de mots (Bag of Words) :** Ces modèles représentent chaque document comme un vecteur de fréquences de mots. Ils sont simples mais ne capturent pas les relations entre les mots.

2. **Word Embeddings (Word2Vec, GloVe, FastText) :** Ces modèles représentent les mots sous forme de vecteurs denses dans un espace vectoriel de dimensions inférieures. Ils capturent les similarités sémantiques entre les mots.

3. **Réseaux de neurones récurrents (RNN) :** Ces modèles sont conçus pour traiter des données séquentielles, comme du texte. Les RNN sont capables de prendre en compte le contexte précédent pour comprendre le sens du texte.

4. **Transformers :** Les transformers sont des architectures de réseau neuronal qui se sont révélées très efficaces pour de nombreuses tâches en NLP. Ils sont capables de capturer les relations à longue distance entre les mots.

5. **Modèles de langue pré-entraînés (BERT, GPT, etc.) :** Ces modèles sont entraînés sur de grandes quantités de texte non étiqueté et peuvent être fine-tunés pour des tâches spécifiques en NLP, comme la classification de texte, la génération de texte, etc.

En résumé, un modèle en NLP est une représentation mathématique d'un système capable de comprendre et de générer du langage humain. Ces modèles sont entraînés sur de grandes quantités de données textuelles et sont capables d'effectuer une variété de tâches en NLP.

Chargement d'un modèle pré-entraîné : Maintenant que nous savons quel modèle utiliser, utilisons-le en Python. Tout d'abord, nous devons importer les classes AutoTokenizer et AutoModelForSequenceClassification de transformers.

L'utilisation de ces classes AutoModel permettra de déduire automatiquement l'architecture du modèle à partir de son nom.

````python
#Le code importe un tokenizer pré-entraîné à partir de la bibliothèque transformers.
#AutoTokenizer est une classe qui donne accès à tous les tokenizers.
#from_pretrained(nom_du_modèle) est une méthode qui charge un tokenizer pré-entraîné
# spécifié par nom_du_modèle.


from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "lxyuan/distilbert-base-multilingual-cased-sentiments-student"

# We call define a model object
model = AutoModelForSequenceClassification.from_pretrained(model_name)
````

````python

# il faut savoir qu'en général lorseque vous chargez un modèle depuis Hugging Face
#Vous le chargez avec son tokenizers
# le tokenizers ici va convertir les mots en réprésentation vectorielle numérique spécifique au modèle
tokenizer = AutoTokenizer.from_pretrained(model_name)
````

Finetuner les modèles

Le fine-tuning est le processus qui consiste à prendre un modèle pré-entraîné et à mettre à jour ses paramètres en l'entraînant sur un ensemble de données spécifique à votre tâche. Cela vous permet de tirer parti des représentations apprises par le modèle et de les adapter à votre cas d'utilisation.

Imaginons que nous devons utiliser un modèle de classification de texte pour déduire les sentiments à partir d'une liste de tweets. Une question naturelle qui vient à l'esprit est la suivante : Ce modèle pré-entraîné fonctionnera-t-il correctement ?

Pour nous en assurer, nous pouvons tirer parti du fine-tuning en entraînant un modèle Hugging Face pré-entraîné avec un ensemble de données contenant des tweets et leurs sentiments correspondants afin d'améliorer ses performances.

2. Préparez votre jeu de données : Maintenant que nous avons déjà notre jeu de données, nous avons besoin d'un tokenizer pour le préparer à être analysé par notre modèle. La variable texte de notre jeu de données doit être tokenisée afin que nous puissions l'utiliser pour affiner notre modèle.

C'est pourquoi la deuxième étape consiste à charger un tokenizer pré-entraîné et à tokeniser notre jeu de données afin qu'il puisse être utilisé pour l'affinement.

````python
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
````

3. Construire un ensemble de données PyTorch avec des encodages : La troisième étape consiste à générer un ensemble de données d'entraînement et de test. L'ensemble d'entraînement sera utilisé pour affiner notre modèle, tandis que l'ensemble de test sera utilisé pour l'évaluer.

Généralement, le processus d'affinage prend beaucoup de temps. Afin de faciliter le tutoriel, nous échantillonnons de manière aléatoire les deux ensembles de données pour réduire le temps de calcul.

````python
from datasets import load_dataset

model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Loading the dataset to train our model
dataset = load_dataset("mteb/tweet_sentiment_extraction")

small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
````

Affinez (finetuner) le modèle :

Notre dernière étape consiste à configurer les arguments d'entraînement et à démarrer le processus d'entraînement. La bibliothèque transformers contient la classe trainer(), qui gère tout.

Nous définissons d'abord les arguments d'entraînement ainsi que la stratégie d'évaluation. Une fois que tout est défini, nous pouvons facilement entraîner le modèle avec la commande train().

````python
from transformers import Trainer, TrainingArguments
import numpy as np
training_args = TrainingArguments(output_dir="trainer_output", evaluation_strategy="epoch")

metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
trainer.train()
````

Ce code est utilisé pour entraîner un modèle en utilisant la bibliothèque `transformers` de Hugging Face. Voici ce que fait chaque partie du code :

1. **Import des bibliothèques nécessaires :**
   ```python
   from transformers import Trainer, TrainingArguments
   import numpy as np
   ```

2. **Définition des arguments d'entraînement :**
   ```python
   training_args = TrainingArguments(output_dir="trainer_output", evaluation_strategy="epoch")
   ```
   - `output_dir`: Le répertoire où les modèles entraînés et les autres fichiers de sortie seront sauvegardés.
   - `evaluation_strategy`: La stratégie d'évaluation. Ici, "epoch" signifie que l'évaluation sera effectuée à la fin de chaque époque.

3. **Définition de la métrique d'évaluation :**
   ```python
   metric = evaluate.load("accuracy")
   ```
   - `evaluate.load("accuracy")` charge la métrique d'exactitude ("accuracy") à partir de la bibliothèque d'évaluation.

4. **Définition de la fonction pour calculer les métriques :**
   ```python
   def compute_metrics(eval_pred):
       logits, labels = eval_pred
       predictions = np.argmax(logits, axis=-1)
       return metric.compute(predictions=predictions, references=labels)
   ```
   - Cette fonction prend les prédictions du modèle (`logits`) et les étiquettes réelles (`labels`) comme entrée.
   - En utilisant `np.argmax`, elle obtient les indices des classes prédites avec les probabilités maximales.
   - Ensuite, elle utilise la métrique d'exactitude pour calculer la performance du modèle.

5. **Configuration de l'entraîneur (trainer) :**
   ```python
   trainer = Trainer(
       model=model,
       args=training_args,
       train_dataset=small_train_dataset,
       eval_dataset=small_eval_dataset,
       compute_metrics=compute_metrics,
   )
   ```
   - `model`: Le modèle que vous souhaitez entraîner.
   - `args`: Les arguments d'entraînement définis précédemment.
   - `train_dataset` et `eval_dataset`: Les ensembles de données d'entraînement et d'évaluation.
   - `compute_metrics`: La fonction pour calculer les métriques.

6. **Entraînement du modèle :**
   ```python
   trainer.train()
   ```
   - Cette commande lance le processus d'entraînement du modèle en utilisant les paramètres spécifiés.

En résumé, ce code configure et entraîne un modèle en utilisant la bibliothèque `transformers`, en spécifiant les arguments d'entraînement, les ensembles de données, et les métriques à utiliser pour évaluer le modèle.
